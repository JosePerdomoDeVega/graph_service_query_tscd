```python
def body_with_correct_format(body: dict):
    correct_format = False
    attributes = set(body.keys())

    expected_attributes = {'callback_url', 'max_depth', 'origin_word', 'storage_key', 'degree', 'job_id', 'operation', 'destination_word'}
    expected_types = {"job_id": str, "operation": str, "storage_key": str or None, "origin_word": str, "destination_word": str, "max_depth": str, "degree": str, "callback_url": dict, "result": dict, "errors": dict}

    if attributes == expected_attributes:
        correct_format = True
        for key in expected_types.keys():
            print(f"VariableActual: {type(body.get(key))}, Expected: {expected_types.get(key)}, Res: {(type(body.get(key)) == expected_types[key])}")
            correct_format = correct_format and (type(body.get(key)) == expected_types[key])

    return correct_format

body_with_correct_format({'job_id': '0bf2d629-f91c-4471-a0e4-8d7c5220172b',
 'storage_key': None,
 'operation': 'min_path',
 'origin_word': 'dog',
 'destination_word': 'aah',
 'max_depth': 'None',
 'degree': 'None',
 'callback_url': 'http://localhost:9000/callback'})
```
